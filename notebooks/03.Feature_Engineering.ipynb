## Import Libraries

```python
import pandas as pd
import numpy as np
```

## Load Dataset

```python
df = pd.read_csv("C:\\Users\\Hey!\\OneDrive\\Desktop\\Customer churn prediction & retention Platform\\processed_data\\telecom_customer_churn_cleaned.csv")

print(df.shape)

df.head()
```

```python
## Check Target Variable

df["Churn Value"].value_counts()
```

```python
## Remove Leakage Columns

drop_cols = [
    "CustomerID","Country","State","City","Zip Code",
    "Latitude","Longitude","Churn Label","Churn Reason"]

df.drop(columns=drop_cols,inplace=True)
```

## Investigate Churn Score

```python
df[["Churn Value","Churn Score"]].corr()
```

## Business Features


```python
## Revenue Per Month

df["Revenue_Per_Month"] = (df["Total Charges"] /(df["Tenure Months"] + 1))
```

```python
##  Customer Value
df["Customer_Value"] = (df["Monthly Charges"] *df["Tenure Months"])
```

```python
## Service Count

service_cols = ["Phone Service","Multiple Lines","Online Security","Online Backup",
                "Device Protection","Tech Support","Streaming TV","Streaming Movies"]

df["Service_Count"] = (df[service_cols].eq("Yes").sum(axis=1))
```

## Encode Categorical Variables

```python
## Check categorical columns:

cat_cols = df.select_dtypes(include="object").columns

print(cat_cols)
```

```python
## Use One-Hot Encoding:

df = pd.get_dummies(df,columns=cat_cols,drop_first=True)
```
