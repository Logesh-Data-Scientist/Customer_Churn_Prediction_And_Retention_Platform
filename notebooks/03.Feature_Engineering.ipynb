## Import Libraries

```python
import pandas as pd
import numpy as np
```

## Load Dataset

```python
df = pd.read_csv("C:\\Users\\Hey!\\OneDrive\\Desktop\\Customer churn prediction & retention Platform\\processed_data\\telecom_customer_churn_cleaned.csv")

print(df.shape)

df.head()
```

```python
df.columns
```

```python
## Check Target Variable

df["Churn Value"].value_counts()
```

```python
# Remove unnecessary columns and leakage columns

drop_cols = [
    "CustomerID","Country","State","City","Zip Code","Latitude",
    "Longitude","Churn Label","Churn Reason","Churn Score","CLTV"]

df.drop(columns=drop_cols, inplace=True)

print(df.shape)
print(df.columns.tolist())
```

## Encode Categorical Variables

```python
## Check categorical columns:

cat_cols = df.select_dtypes(include="object").columns

print(cat_cols)
```

```python
## Use One-Hot Encoding:

df = pd.get_dummies(df,columns=cat_cols,drop_first=True)
```

## Check Dataset

```python
print(df.shape)
print(df.columns)
```
